```python
df = pd.read_excel('data offstat.xlsx')
df.head()
```

```python
df = df.rename(columns={
    'IPM (Indeks \nPembangunan Manusia)': 'IPM',
    'Kepadatan Penduduk \nper km persegi (Km2)': 'Kepadatan',
    'Tingkat Pengangguran Terbuka (TPT) (Persen)': 'TPT',
    'Persentase Penduduk Miskin  (Persen)': 'Kemiskinan',
    'Luas Panen \nTanaman Padi \n(ha) (Ha)': 'Luas_Panen',
    'Produktivitas \nTanaman Padi \n(ku/ha) (Ku/ha)': 'Produktivitas',
    'Rekap Produksi \nPadi \n(ton) (Ton)': 'Produksi_Padi',
    'Prevalensi Bayi Gizi Kurang (%)': 'Gizi_Kurang'
})
```

```python
df.isna().sum()
```

## Statistika Deskriptif

```python
vars_num = [
    'IKP',
    'IPM',
    'Kepadatan',
    'TPT',
    'Kemiskinan',
    'Luas_Panen',
    'Produktivitas',
    'Produksi_Padi',
    'Gizi_Kurang'
]
```

```python
desc = pd.DataFrame({
    'Mean': df[vars_num].mean(),
    'Median': df[vars_num].median(),
    'Std Dev': df[vars_num].std(),
    'Minimum': df[vars_num].min(),
    'Maximum': df[vars_num].max()
})

desc
```

```python
desc.to_excel('Statistik_Deskriptif.xlsx')
```

## Histogram

```python
hist_vars = [
    'IKP',
    'IPM',
    'Kemiskinan',
    'Gizi_Kurang'
]
```

```python
for col in hist_vars:

    plt.figure(figsize=(6,4))

    sns.histplot(
        df[col],
        bins=10,
        kde=True
    )

    plt.title(f'Histogram {col}')
    plt.show()
```

## Boxplot

```python
for col in vars_num:

    plt.figure(figsize=(6,3))

    sns.boxplot(
        x=df[col]
    )

    plt.title(f'Boxplot {col}')

    plt.show()
```

## Correlation Matrix

```python
corr = df[vars_num].corr()

corr
```

```python
corr.to_excel('Correlation_Matrix.xlsx')
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    corr,
    annot=True,
    cmap='RdBu_r',
    fmt='.2f'
)

plt.title('Heatmap Korelasi')

plt.show()
```

## Korelasi IKP dengan Variabel lain

```python
ikp_corr = corr['IKP'].sort_values(
    ascending=False
)

ikp_corr
```

```python
ikp_corr = ikp_corr.drop('IKP')

ikp_corr
```

```python
ikp_table = pd.DataFrame({
    'Variabel': ikp_corr.index,
    'Korelasi dengan IKP': ikp_corr.values
})

ikp_table = ikp_table.reindex(
    ikp_table['Korelasi dengan IKP'].abs().sort_values(ascending=False).index
)

ikp_table
```

```python
plt.figure(figsize=(8,5))

ikp_corr.sort_values().plot(
    kind='barh'
)

plt.title(
    'Korelasi Variabel terhadap IKP'
)

plt.xlabel('Koefisien Korelasi')

plt.show()
```

## Top 5 Korelasi Terkuat dengan IKP

```python
ikp_abs = ikp_corr.abs().sort_values(
    ascending=False
)

print(
    ikp_abs.head()
)
```
